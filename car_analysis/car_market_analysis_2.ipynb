### Страна-производитель

```python
data['manufacturing_country'].unique()
```

Кроме UK все остальные страны прописаны на русском языке. Можно закодировать в Alpha3, для этого нужно сначала перевести страны на английский. Чтобы не делать перевод для каждой строки, не обращаться почти 40 тысяч раз в переводчик, мы создадим список уникальных строк, переведем их и вставим обратно в наш датасет.

```python
# список наших стран
country_rus=data['manufacturing_country'].unique().tolist()
translator = Translator()
# список стран на английском
country = [translator.translate(country, src='ru', dest='en').text for country in country_rus]
# датасет с двумя колонками: русское и английское название
df = pd.DataFrame({'Русское': country_rus, 'Английское': country})
# функция для замены в датасете
def replace_country(row):
    # Ищем соответствие в df
    match = df[df['Русское'] == row['manufacturing_country']]
    if not match.empty:
        return match.iloc[0]['Английское']
    else:
        return row['manufacturing_country']  # Возвращаем исходное название, если соответствие не найдено
# применяем функцию ко всем строкам в data
data['manufacturing_country'] = data.apply(replace_country, axis=1)
# проверяем, что получилось
data['manufacturing_country'].unique()
```

Перевод выполнен, теперь перейдем к кодированию

```python
# функция кодирования
def get_alpha3_code(country_name):
    try:
        country_obj = pycountry.countries.get(name=country_name)
        return country_obj.alpha_3 if country_obj else country_name
    except:
        return country_name  # Возвращаем исходное название, если страна не найдена

# применяем функцию к столбцу
data['manufacturing_country'] = data['manufacturing_country'].apply(get_alpha3_code)
```

```python
data['manufacturing_country'].unique()
```

```python
def fill_missing_fuel(df):
    # Группируем по бренду, моделе
    group_columns = ['brand', 'model']  

    # Заполняем пропущенные значения наиболее частым значением в группе
    df['fuel_type'] = df.groupby(group_columns)['fuel_type'].transform(
        lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x)
    )
    df['engine_volume'] = df.groupby(group_columns)['engine_volume'].transform(
        lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x)
    )
    df['transmission_type'] = df.groupby(group_columns)['transmission_type'].transform(
        lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x)
    )
    df['drive_type'] = df.groupby(group_columns)['drive_type'].transform(
        lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x)
    )
    return df
```

```python
def categorize_column(input_column):

    conditions = [
        (input_column < 1.2),
        (input_column >= 1.2) & (input_column < 1.8),
        (input_column >= 1.8) & (input_column < 3.5),
        (input_column >= 3.5)
    ]

    # Пример категорий. Замените их на свои.
    categories = ['Малый', 'Средний', 'Большой','Огромный']

    # Применение условий и создание нового столбца с категориями
    output_column = pd.cut(input_column, bins=[float('-inf'), 1.2,1.8,3.5, float('inf')], labels=categories)

    return output_column

data['cat_vol'] = categorize_column(data['engine_volume'])
```

```python
# создаем таблицу с сегментом и объемом двигателем
vol_cars=data.loc[data['segment_2013']!='Коммерческие автомобили'].pivot_table(index='segment_2013', columns='cat_vol', values='sell_price_usd', aggfunc='sum')\
.rename_axis(None, axis=1)
```

Получается, что больше всего денег в Казахстане в 2019 году дилерам приносил - бензиновый внедорожник с объемом двигателя от 1.8 до 3.5 литров. Можем посмотреть количество продаваемых автомобилей.

### RFM-анализ моделей авто (показатели Frequency и Monetary)

```python
# считаем значения частоты покупок и денежной стоимости
data['f_value'] = data.groupby('model')['model'].transform('count')
data['m_value'] = data.groupby('model')['sell_price_usd'].transform('sum')
```

```python
# сегментируем по частоте покупок и денежной стоимости
data['f'] = pd.qcut(data['f_value'], 3, labels=[1, 2, 3])
data['m'] = pd.qcut(data['m_value'], 3, labels=[1, 2, 3])

# выделяем сегменты
data['fm'] = data['f'].astype('str') + data['m'].astype('str')
```

Описание значений сегментов : 
* Monetary :  1 - низкие продажи,  2 - средние продажи,        3 - высокие продажи
* Frequency:  1 - редко продаются, 2 - средняя частота продаж, 3 - часто продаются
    

Значит, самый лучший сегмент 33 - часто продают по высокой цене, интересные для нас сегменты со значениями в Frequency и Monetary 2 или 3. Выведем ждя этиих сегментов модели автомобилей.
