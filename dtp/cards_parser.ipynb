```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse
```

```python


# URL страницы, где находятся файлы
page_url = "https://cms.dtp-stat.ru/media/opendata/"

# Папка для сохранения скачанных файлов
save_folder = "/Users/juliagrebeneva/Desktop/cardssss"

# Создание папки, если она не существует
if not os.path.exists(save_folder):
    os.makedirs(save_folder)

# Отправляем GET запрос для получения HTML содержимого страницы
response = requests.get(page_url)

# Проверяем успешность запроса
if response.status_code == 200:
    # Инициализируем парсер BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Извлекаем ссылки на файлы
    file_links = soup.find_all("a", href=True)
    
    # Скачиваем каждый файл
    for link in file_links:
        # Получаем URL файла
        file_url = page_url + link["href"]
        
        # Извлекаем имя файла из URL
        file_name = os.path.basename(urlparse(file_url).path)
        
        # Определяем полный путь для сохранения файла
        file_path = os.path.join(save_folder, file_name)
        
        # Проверяем, существует ли файл
        if not os.path.exists(file_path):
            # Отправляем GET запрос для скачивания файла
            file_response = requests.get(file_url)
            
            # Проверяем успешность запроса скачивания файла
            if file_response.status_code == 200:
                # Сохраняем содержимое файла
                with open(file_path, "wb") as file:
                    file.write(file_response.content)
                print(f"Файл {file_name} успешно загружен.")
            else:
                print(f"Ошибка загрузки файла {file_url}: {file_response.status_code}")
        else:
            print(f"Файл {file_name} уже существует.")
else:
    print(f"Ошибка получения HTML содержимого страницы: {response.status_code}")
```
